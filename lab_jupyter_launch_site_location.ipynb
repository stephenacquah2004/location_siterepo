```python
# Task 2: Mark the success/failed launches for each site on the map
# Create a map object
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Iterate through each launch record in spacex_df
for index, row in spacex_df.iterrows():
    # Get the coordinates of the launch site
    site_coords = (row['Lat'], row['Long'])
    
    # Determine marker color based on launch outcome
    marker_color = 'green' if row['class'] == 1 else 'red'
    
    # Add marker to the map
    folium.Marker(site_coords, 
                  icon=folium.Icon(color=marker_color), 
                  popup=f"Launch Site: {row['Launch Site']}, Outcome: {'Success' if row['class'] == 1 else 'Failure'}"
                 ).add_to(site_map)

# Display the map
site_map
```

Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
# Create a MarkerCluster object
marker_cluster = MarkerCluster()

# Iterate through each row in spacex_df
for index, row in spacex_df.iterrows():
    # Determine the marker color based on the 'class' column
    marker_color = 'green' if row['class'] == 1 else 'red'
    
    # Create a Marker object for each launch record and add it to the MarkerCluster
    folium.Marker(
        location=[row['Lat'], row['Long']],  # Coordinates for the launch site
        popup=row['Launch Site'],  # Popup label showing the launch site
        icon=folium.Icon(color=marker_color)  # Marker color based on success/failure
    ).add_to(marker_cluster)

# Add the MarkerCluster to the site map
site_map.add_child(marker_cluster)

```

*TODO:* Create a new column in `launch_sites` dataframe called `marker_color` to store the marker colors based on the `class` value


```python
# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
# Create a new column 'marker_color' based on the 'class' column

# Apply a function to check the value of 'class' column and assign marker colors
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')
```

```python
from geopy.distance import geodesic

# Define coordinates for launch sites
lat1, long1 = 29.559684888503615, -95.0830971930759  
lat2, long2 = 28.5721, -80.648  

# Define coordinates for nearby locations/features
latA, longA = 40.7128, -74.0060  
latB, longB = 37.7749, -122.4194  

launch_site_coordinates = {
    'Launch Site 1': (lat1, long1),
    'Launch Site 2': (lat2, long2),
}
nearby_location_coordinates = {
    'City A': (latA, longA),
    'Airport B': (latB, longB),
}

distances = {}
for launch_site, launch_site_coord in launch_site_coordinates.items():
    distances[launch_site] = {}
    for location, location_coord in nearby_location_coordinates.items():
        distances[launch_site][location] = geodesic(launch_site_coord, location_coord).kilometers

for launch_site, location_distances in distances.items():
    print(f"Distances from {launch_site}:")
    for location, distance in location_distances.items():
        print(f"{location}: {distance:.2f} kilometers")
```

Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Add MousePosition plugin to the map
mouse_position = MousePosition(position='topright', separator=' | ')
site_map.add_child(mouse_position)

# Once you click on the map to mark down a point on the closest coastline,
# read the coordinates from the MousePosition display
# Assign the coordinates to variables coastline_lat and coastline_lon
# Example:
coastline_lat =  28.56367
coastline_lon = --80.57163

# Use the launch site coordinates from launch_sites_df instead of launch_site_lat and launch_site_lon
launch_site_coords = (launch_sites_df.iloc[0]['Lat'], launch_sites_df.iloc[0]['Long'])

# Calculate distance between launch site and marked coastline point
distance_coastline = calculate_distance(launch_site_coords[0], launch_site_coords[1], coastline_lat, coastline_lon)
print("Distance to closest coastline:", distance_coastline, "km")
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example
# distance_marker = folium.Marker(
#    coordinate,
#    icon=DivIcon(
#        icon_size=(20,20),
#        icon_anchor=(0,0),
#        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
#        )
#    )
# Create and add a folium.Marker on the selected closest coastline point on the map
distance_marker = folium.Marker(
    [coastline_lat, coastline_lon],  # Coordinates of the closest coastline point
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline),
    )
)

# Add the distance_marker to the map
site_map.add_child(distance_marker)
```
